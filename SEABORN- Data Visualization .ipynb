```python
# IMPORTING NECESSARY LIBRARAIES
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# WE WILL USE IRIS DATASET
df=pd.read_csv('iris.csv')
df
```

```python
df.head(5)# TO DISPLAY FIRST FIVE ROWS
```

```python
# FOR SCATTER PLOT
sns.lmplot(x="sepal_length", y ="petal_length", fit_reg=False, hue="sepal_width", data=df)# fit_reg-- to remove regression line
#plt.ylim(0, None) these for when axes fall below zero, here graph is already above zero so no need to use it .
#plt.xlim(0, None)
```

```python
# BOX PLOT
plt.figure(figsize=(10,6))
sns.set_style('darkgrid')
sns.boxplot(data=df)
plt.show()
```

```python
#CATPLOT
sns.catplot(data=df)
#SETTING LABELS AND TITLE WITH MATPLOTLIB
plt.xlabel('Features')
plt.ylabel('Frequency')
plt.title(" Iris data Plot")
plt.show
```

```python
#VIOLIN PLOT
plt.figure(figsize=(10,6))
sns.violinplot(x= "sepal_length", y= "petal_length", width= 6.0,hue="Class",  inner=None, data=df)#  INNER= NoNE, to remove bars inside violin
plt.show()
```

```python
# BAR PLOT
plt.figure(figsize=(10,6))
sns.barplot(x="sepal_length", y="sepal_width", data=df, ci=95,n_boot=1000 )
plt.show()
```

```python
# SWARM PLOT to display each point
plt.figure(figsize=(10,6))
sns.swarmplot(x="sepal_length", y="sepal_width", data=df,alpha=1.0)
plt.show
```

```python
#NOW COMBINING SWARM PLOT AND VIOLIN PLOT AS THEY DISPLAY SIMILAR INFO.
plt.figure(figsize=(10,6))
sns.swarmplot(x="sepal_length", y="sepal_width", data=df,alpha=0.6, color="Black")# alpha = make points transparent
sns.violinplot(x="sepal_length", y="sepal_width", data=df, width=6.0, inner=None)# inner=None for removing bars inside violin
# Set title with matplotlib
plt.title("Sepal Length  and sepal width")
plt.show()
```
